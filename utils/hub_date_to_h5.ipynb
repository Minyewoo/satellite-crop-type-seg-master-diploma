```python
import os

dataset_name = 'dlr_fusion_competition_germany'
sentinel2_name = f'{dataset_name}_train_source_sentinel_2'
labels_name = f'{dataset_name}_train_labels'
download_path = os.path.join('.', 'germany_sentinel2_data')

data_coords = '33N_18E_242N'
geojson_dir = os.path.join(download_path, labels_name, f'{labels_name}_{data_coords}')
bands_dir = os.path.join(download_path, sentinel2_name, f'{sentinel2_name}_{data_coords}_2018')
```

```python
import tarfile
from glob import glob

for gz in glob(os.path.join(download_path, "*.gz")):
    #unpacking
    print(f"unzipping {gz} to {download_path}")
    tar = tarfile.open(gz, "r:gz")
    tar.extractall(download_path)
    tar.close()
    #cleaning up
    os.remove(gz)
```

```python
import tables
import h5py as h5
import h5py_cache as h5c
import numpy as np

#merging 2 datasets into one
hdf5_path = 'sentinel_2_dataset.h5'
f = h5c.File(hdf5_path, 'w',rdcc_nbytes=1024**2*4000, rdcc_nslots=1e7)
dset = f.create_dataset('sentinel2_data', shape=(220, 2400, 2400, 12), dtype=np.float32, chunks=(1, 300, 300, 1), compression='lzf')

germ_bands = np.load(os.path.join(bands_dir, 'bands.npy'), mmap_mode='r')
for i in range(germ_bands.shape[0]):
  print(i)
  dset[i] = germ_bands[i]

f.close()
```
